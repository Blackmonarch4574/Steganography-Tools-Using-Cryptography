```python


import numpy as np
import cv2
from matplotlib import pyplot as plt
from encode_txt_data import encode_txt_data
from decode_txt_data import decode_txt_data

from encode_img_data import encode_img_data
from decode_img_data import decode_img_data

from encode_aud_data import encode_aud_data
from decode_aud_data import decode_aud_data

from encode_vid_data import encode_vid_data
from decode_vid_data import decode_vid_data

import tkinter as tk
from tkinter import messagebox

```

```python
def txt_steg():
    while True:
        print("\n\t\tTEXT STEGANOGRAPHY OPERATIONS") 
        print("1. Encode the Text message")  
        print("2. Decode the Text message")  
        print("3. Exit")  
        choice1 = int(input("Enter the Choice:"))   
        if choice1 == 1:
            encode_txt_data()
        elif choice1 == 2:
            decrypted=decode_txt_data() 
        elif choice1 == 3:
            break
        else:
            print("Incorrect Choice")
        print("\n")
        messagebox.showinfo("Text Steganography", "This is the Text Steganography function.")
        
```

```python

def img_steg():
    while True:
        print("\n\t\tIMAGE STEGANOGRAPHY OPERATIONS\n") 
        print("1. Encode the Text message") 
        print("2. Decode the Text message") 
        print("3. Exit")  
        choice1 = int(input("Enter the Choice: "))   
        if choice1 == 1:
            image=cv2.imread("Sample_cover_files/cover_image.jpg")
            encode_img_data(image)
        elif choice1 == 2:
            image1=cv2.imread(input("Enter the Image you need to Decode to get the Secret message :  "))
            decode_img_data(image1)
        elif choice1 == 3:
            break
        else:
            print("Incorrect Choice")
        print("\n")
        messagebox.showinfo("Image Steganography", "This is the Image Steganography function.")
```

```python

def aud_steg():
    while True:
        print("\n\t\tAUDIO STEGANOGRAPHY OPERATIONS") 
        print("1. Encode the Text message")  
        print("2. Decode the Text message")  
        print("3. Exit")  
        choice1 = int(input("Enter the Choice:"))   
        if choice1 == 1:
            encode_aud_data()
        elif choice1 == 2:
            decode_aud_data()
        elif choice1 == 3:
            break
        else:
            print("Incorrect Choice")
        print("\n")
        messagebox.showinfo("Audio Steganography", "This is the Audio Steganography function.")
```

```python

def vid_steg():
    while True:
        print("\n\t\tVIDEO STEGANOGRAPHY OPERATIONS") 
        print("1. Encode the Text message")  
        print("2. Decode the Text message")  
        print("3. Exit")  
        choice1 = int(input("Enter the Choice:"))   
        if choice1 == 1:
            a=encode_vid_data()
        elif choice1 == 2:
            decode_vid_data(a)
        elif choice1 == 3:
            break
        else:
            print("Incorrect Choice")
        print("\n")
        messagebox.showinfo("Video Steganography", "This is the Video Steganography function.")
```

```python
def main():
    root = tk.Tk()
    root.title("Steganography App")

    label = tk.Label(root, text="STEGANOGRAPHY", font=("Helvetica", 16, "bold"))
    label.pack(pady=10)

    label_menu = tk.Label(root, text="MAIN MENU", font=("Helvetica", 14, "bold"))
    label_menu.pack()

    options = [
        "TEXT STEGANOGRAPHY {Hiding Text in Text cover file}",
        "IMAGE STEGANOGRAPHY {Hiding Text in Image cover file}",
        "AUDIO STEGANOGRAPHY {Hiding Text in Audio cover file}",
        "VIDEO STEGANOGRAPHY {Hiding Text in Video cover file}",
        "Exit"
    ]

    choice_var = tk.IntVar()
    choice_var.set(0)

    for i, option in enumerate(options, start=1):
        tk.Radiobutton(root, text=option, padx=20, variable=choice_var, value=i).pack(anchor="w")

    def execute_choice():
        choice = choice_var.get()
        if choice == 1:
            txt_steg()
        elif choice == 2:
            img_steg()
        elif choice == 3:
            aud_steg()
        elif choice == 4:
            vid_steg()
        elif choice == 5:
            root.destroy()  # Exit the GUI
        else:
            messagebox.showwarning("Incorrect Choice", "Please select a valid option.")

    btn_execute = tk.Button(root, text="Execute", command=execute_choice, bg="green", fg="white")
    btn_execute.pack(pady=10)

    root.mainloop()

if __name__ == "__main__":
    main()
```
